```python
print("Venkat Chalamalla 21BDS0276")
kmeans = KMeans(n_clusters=3)
data['cluster'] = kmeans.fit_predict(data[['abil', 'educ', 'exper']])
sns.pairplot(data, vars=['abil', 'educ', 'exper'], hue='cluster', palette='viridis')
fig = plt.figure(figsize=(6, 6))
plt.show()
```

```python
print("Venkat Chalamalla 21BDS0276")
X = data[['educ', 'abil', 'exper', 'urban']]
y = data['wage']
model = LinearRegression()
model.fit(X, y)
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("R-squared:", model.score(X, y))
```
